```python
Q1 = df.drop(['Unnamed: 0','Cow','Date'],axis=1).quantile(0.25)
Q3 = df.drop(['Unnamed: 0','Cow','Date'],axis=1).quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
sns.boxplot(x=df['Body Weight (kg)'])
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(x=df['Daily Hay Dry matter (kg)'])
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(x=df['hay intake fresh'])
```

```python
import matplotlib.pyplot as plt
# Calculate the correlation matrix
corr = df.drop(['Unnamed: 0','Cow','Date'],axis=1).corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
